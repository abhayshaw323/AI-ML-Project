```python
# Given dataset
data = ["red", "blue", "?", "green", "blue", "", "N/A", "green", "blue"]

# List of invalid values
invalid_values = ["?", "N/A", "None", ""]

# Step 1: Create a new list with only valid categories
clean_data = []
for item in data:
    if item not in invalid_values:
        clean_data.append(item)

# Step 2: Manually compute frequency using loops (no Counter)
frequency = {}
for item in clean_data:
    if item in frequency:
        frequency[item] += 1
    else:
        frequency[item] = 1

# Step 3: Find the category with the highest frequency
max_category = None
max_count = 0

for category in frequency:
    if frequency[category] > max_count:
        max_count = frequency[category]
        max_category = category

print("Cleaned Data:", clean_data)
print("Frequency Count:", frequency)
print(f"Most Frequent Category: {max_category} ({max_count} times)")
```

```python
# Given data
temps = [30, 32, 31, 29, 28, 35, 36]
k = 3  # Window size

# List to store moving averages
moving_avg = []

# Loop through valid windows
for i in range(len(temps) - k + 1):
    window_sum = 0
    
    # Calculate sum of current window
    for j in range(k):
        window_sum += temps[i + j]
    
    # Compute average and append to list
    avg = window_sum / k
    moving_avg.append(avg)

print("Moving Averages:", moving_avg)
```
